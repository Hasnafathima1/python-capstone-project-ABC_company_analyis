```python
#import python libraries 
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
#read the csv file
df=pd.read_csv('ABC Company.csv')
print(df.head())
```

```python
#1.preprocessing data 
#Correct the data in the "height" column by replacing it with random
#numbers between 150 and 180.
df['height']=np.random.randint(150,180,len(df))
print(df)
```

```python
#2.Analising task 
#(A)distribution employes across teams
team_counts=df['Team'].value_counts()#count employess per team
print(team_counts)
#the team with most employes is "New orleans pelicans"
```

```python
#percentage of employess per team
total_employes=len(df)
total_percent=(team_counts/total_employes)*100
print(total_percent)
print("Total employes:",total_employes)#total employes
```

```python
#barchart of "counts of employes per team"
team_counts.plot(color="yellow",marker="o",figsize=(6,4))
plt.title('counts of employes per team')
plt.xlabel('team')
plt.ylabel('number of employes')
plt.show
#the team with most employes is "New orleans pelicans "
```

```python
#(B)segregate employes across based on thier position
team_position=df["Position"].value_counts()
print(team_position)
```

```python
#bar chart visualization
team_position.plot(kind="bar",color="blue")
plt.title("Employes by position")
plt.xlabel("position")
plt.ylabel("count")
plt.show()
#highest employes are in "SG" comparetively others 
```

```python
#(C)predominant age group
bins=[20,30,40,50,60,70]
labels=['21-30','31-40','41-50','51-60','61-70']


#put employes in age group
df['Age group']=pd.cut(df['Age'],bins=bins,labels=labels

                       #counts each group
age_group_counts=df['Age group'].value_counts()
print(age_group_counts)
```

```python
#bar chart visualization
age_group_counts.plot(kind="bar",color='green')
plt.title("Employes by Age")
plt.xlabel("Age Group")
plt.ylabel("Number of employes")
plt.show()
           #the age group  '21-30' consist of large number of employes
```

```python
#(D)calculate total salary by team 
salary_by_team=df.groupby('Team')['Salary'].sum()
#add all the salary value for each team
print(salary_by_team)
```

```python
#calculate salary by position
salary_by_position=df.groupby('Position')['Salary'].sum()
print(salary_by_position)
```

```python
#barchart 
salary_by_team.plot(kind='bar',color='red')
plt.title("Salary by Team")
plt.ylabel("Total salary")
plt.show()
#team -'cleveland cavaliers' is the highest salary ependiture in team
```

```python
salary_by_position.plot(kind='bar',color='black')
plt.title("Salary by Position")
plt.ylabel("Total salary")
plt.show()
#position-PG is the slightly more salary ependiture other positions 
```

```python
#(E)correlation between age and salary
corr=df['Age'].corr(df['Salary'])
print("correlation between age and salary:",corr)
#correlation is around 0.20
#The relationship between Age and Salary is weak positive.

#That means older employees earn slightly more, but age is not a strong predictor of salary.
#→ As age increases, salary tends to increase a little bit.
#→ But it’s not strong — some young people earn more than older ones.
```

```python


# Scatter plot
plt.scatter(df['Age'], df['Salary'], alpha=0.5)
plt.title(f"Age vs Salary (Correlation = {corr:.2f})")
plt.xlabel("Age")
plt.ylabel("Salary")
plt.show()
```
